```python
import time
import matplotlib.pyplot as plt

def tower_of_hanoi(n, source, destination, auxiliary):
    if n > 0:
        # Move n-1 disks from source to auxiliary peg
        tower_of_hanoi(n-1, source, auxiliary, destination)

        # Move the n-1 disks from auxiliary peg to destination peg
        tower_of_hanoi(n-1, auxiliary, destination, source)

def plot_running_time(n_values, running_times):
    plt.plot(n_values, running_times, 'o-')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Running Time (seconds)')
    plt.title('Running Time of Tower of Hanoi')
    plt.show()

# Test the algorithm and measure the running time
n_values = [10, 20]  # Vary the input size
running_times = []

for n in n_values:
    start_time = time.time()
    tower_of_hanoi(n, 'A', 'C', 'B')
    end_time = time.time()
    running_time = end_time - start_time
    running_times.append(running_time)

# Plot the running time
plot_running_time(n_values, running_times)
```
